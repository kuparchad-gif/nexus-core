```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data_all = pd.read_csv('/home/thomas/Downloads/data(2).csv')

# Group the data by 'Criteria' and create a heatmap for each group
criteria_groups = data_all[['Criteria', 'Model', 'Opponent', 'Won']].groupby('Criteria')

for criteria, group in criteria_groups:
    pivot_table = group.pivot_table(index='Model', columns='Opponent', values='Won', aggfunc='mean')
    plt.figure(figsize=(10, 8))
    sns.heatmap(pivot_table, annot=True, fmt=".2f", cmap="coolwarm", cbar=True)
    plt.title(f"Heatmap of Win Rate for Criteria: {criteria}")
    plt.ylabel('Model')
    plt.xlabel('Opponent')
    plt.show()
```
